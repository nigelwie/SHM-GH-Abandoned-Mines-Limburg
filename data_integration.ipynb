## Data integration

```python
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
```

```python
data_gnss = pd.read_csv('gnss\gnss_data.csv')
data_insar_1 = pd.read_csv('insar\sentinel1_asc_t088_limburg.csv')
data_insar_2 = pd.read_csv('insar\sentinel1_dsc_t037_limburg.csv')
data_levelling = pd.read_csv('levelling\Levelling_NAP_ZL_20220616_redef.csv')
```

```python
start_col = next(i for i, col in enumerate(data_insar_1.columns) if col.startswith('d_'))
# Loop through all columns after 'd_20220713'
for col in data_insar_1.columns[start_col:]:
    if col.startswith('d_'):
        new_col = f"{col}_u"
        data_insar_1[new_col] = data_insar_1[col] / np.cos(data_insar_1['pnt_incidangle'] * np.pi / 180)
```
