```python
import geopandas as gpd
import matplotlib.pyplot as plt

shp_area = "Concessions_SL_area_v1.shp"
gdf_area = gpd.read_file(shp_area)

gdf_area.head()
```

```python
ax = gdf_area.plot(figsize=(10, 10), edgecolor="black", facecolor="none")
ax.set_title("Concessions SL  area")
plt.show()
```

```python
import contextily as ctx

gdf_area_3857 = gdf_area.to_crs(epsg=3857)

ax = gdf_area_3857.plot(
    figsize=(10, 10),
    edgecolor="black",
    facecolor="none",
    linewidth=1
)

ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_axis_off()
ax.set_title("Concessions SL  area + basemap")
plt.show()
```

```python
shp_sl = "Concessions_SL_v1.shp"
gdf_sl = gpd.read_file(shp_sl)

gdf_sl.head()
```

```python
ax = gdf_sl.plot(figsize=(10, 10), edgecolor="black", facecolor="none")
ax.set_title("Concessions SL")
plt.show()
```

```python
gdf_sl_3857 = gdf_sl.to_crs(epsg=3857)

ax = gdf_sl_3857.plot(
    figsize=(10, 10),
    edgecolor="black",
    facecolor="none",
    linewidth=1
)

ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)
ax.set_axis_off()
ax.set_title("Concessions SL + basemap")
plt.show()
```

```python
gdf_area_3857 = gdf_area.to_crs(epsg=3857)
gdf_sl_3857   = gdf_sl.to_crs(epsg=3857)

ax = gdf_area_3857.plot(
    figsize=(10, 10),
    edgecolor="black",
    facecolor="none",
    linewidth=1,
    label="SL area"
)

gdf_sl_3857.plot(
    ax=ax,
    edgecolor="red",
    facecolor="none",
    linewidth=1,
    label="SL"
)

ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)
ax.legend()
ax.set_axis_off()
plt.show()
```
